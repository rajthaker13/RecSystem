```python
df = pd.read_csv("patient_info_CONFIDENTIAL.csv")
```

```python
df2 = pd.read_csv("video_watched_events_CONFIDENTIAL.csv")
```

# Select needed columns from two datasets and merge into a separate file

```python
dfPart = df[["patient_id","age","sex","has_bh_specialist"]]
```

```python
df2Part = df2[["patient_id","url","primary_category","secondary_category","description","tags"]]
```

```python
dfFinal = pd.merge(dfPart,  
                      df2Part,  
                      on ='patient_id',  
                      how ='inner') 
dfFinal
```

```python
dfFinal.to_csv(r'final.csv')
```
